```python
%matplotlib inline

from sympy import Matrix, symbols, init_printing, linsolve
from sympy.plotting import plot, plot3d, plot3d_parametric_line, plot3d_parametric_surface

init_printing()
```

```python
x1, x2, x3, x4 = symbols('x1:5')
x1, x2, x3, x4
```

```python
A = Matrix([
    x1 - 2*x2 + x3, 
    2*x2 - 8*x3 - 8, 
    5*x1 - 5*x3 - 10
])
A
```

```python
A.rref()
```

```python
linsolve(A, (x1, x2, x3))
```

```python
f1 = x1 - 2*x2 + x3
f2 = 2*x2 - 8*x3 - 8
f3 = 5*x1 - 5*x3 - 10


x3_val = -1
f1_fixed = f1.subs(x3, x3_val)
f2_fixed = f2.subs(x3, x3_val)
f3_fixed = f3.subs(x3, x3_val)

f1_fixed, f2_fixed, f3_fixed
```

```python
plot3d(f1_fixed, f2_fixed, f3_fixed, (x1, -5, 5), (x2, -5, 5))
```
