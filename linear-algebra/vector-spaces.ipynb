```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np

import ipywidgets as widgets
from mpl_toolkits.mplot3d import axes3d
```

```python
# define vectors to plot
V = np.array([[1,1], [-2,2], [4,-7]]) # array row vectors to plot
origin = np.array([[0, 0, 0],[0, 0, 0]]) # origin point

# set up axes for plot
fig, ax = plt.subplots(figsize = (4,4))
ax.axis([-3,5,-8,3])
# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# plot vectors
ax.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], angles='xy', scale_units='xy', scale=1)
```

```python
fig, ax = plt.subplots()

soa = np.array([
    [0, 0, 0, -2, 0, 1], 
    [0, 0, 0, 4, 5, 3],
    [0, 0, 0, 6, 5, 2],
    [-2, 0, 1, 6, 5, 2],
    [6, 5, 2, -2, 0, 1],
])

X, Y, Z, A, B, C = zip(*soa)

ax = fig.add_subplot(projection='3d')
ax.set_xlim([-3, 6])
ax.set_ylim([-3, 6])
ax.set_zlim([0, 4])

q = ax.quiver(
    X, Y, Z, A, B, C, 
    color=['r', 'b', 'g', 'k', 'k'],
    linestyle=['-', '-', '-', '--', '--'],
    linewidth=1,
    arrow_length_ratio=0.03, 
)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')

ax.set_xlim([-20, 21])
ax.set_ylim([-20, 21])
ax.set_zlim([-20, 21])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

points = [
    [0, 0, 0, 'k'], # origin
    
    # [1, 0, 2, 'r'],
    # [-3, -2, 4, 'g'],
    # [1, 8, -5, 'b'],
    # [-4, -2, 2, 'k'],

    [5, -6, 7, 'k'],
    [-5, 6, -7, 'k'],
]

x, y, z, c = zip(*points)
ax.scatter(x, y, z, color=c)

axis_len = 20
vectors = [
    # X, Y, Z line
    [0, 0, 0, axis_len, 0, 0, 'r', '-'], # X
    [0, 0, 0, -axis_len, 0, 0, 'r', '--'], # X
    [0, 0, 0, 0, axis_len, 0, 'g', '-'], # Y
    [0, 0, 0, 0, -axis_len, 0, 'g', '--'], # Y
    [0, 0, 0, 0, 0, axis_len, 'b', '-'], # Z
    [0, 0, 0, 0, 0, -axis_len, 'b', '--'], # Z
    
    # [1, 0, 2, -4, -2, 2, 'b', '-'], # A-B, supposedly -3, -2, 2, but it looks wrong
    # [0, 0, 0, -4, -2, 2, 'b', '--'], # O-B

    # [1, 0, 2, 1, 8, -5, 'b', '-'], # A-C
    # [0, 0, 0, 0, 8, -7, 'b', '--'], # O-C

    [0, 0, 0, 5, -6, 7, 'k', '-'],
    [0, 0, 0, -5, 6, -7, 'k', '-'],
]

X, Y, Z, U, V, W, c, ls = zip(*vectors)
q = ax.quiver(
    X, Y, Z, U, V, W, 
    color=c,
    linestyle=ls,
    arrow_length_ratio=0.05, 
)
```
