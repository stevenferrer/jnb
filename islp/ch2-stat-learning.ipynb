```python
fig, ax = subplots(figsize=(6, 6))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f);
```

```python
fig, ax = subplots(figsize=(6, 6))
ax.imshow(f);
```

### 2.3.6 Indexing data

```python
A = np.array(np.arange(16)).reshape((4, 4))
A
```

```python
A[0, 1]
```

#### Indexing rows, columns, and submatrices

Selet multiple rows

```python
A[[1, 3]]
```

Select multiple columns

```python
A[:, [0, 2]]
```

Select a submatrix (2nd and 4th rows, 1st and 3rd columns)

```python
A[[1, 3], :]
```

```python
A[:, [0, 2]]
```

```python
A[[1, 3], [0, 2]]
```

```python
A[[1, 3]][:, [0, 2]]
```

```python
A[np.ix_([1, 3], [0, 2, 3])]
```

#### Boolean indexing

```python
A.shape[0]
```

```python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1, 3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([0, 1, 0, 1]))
```

```python
A[np.array([0, 1, 0, 1])]
```

```python
A[keep_rows]
```

```python
A.shape[1]
```

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
keep_cols
```

```python
idx_bool = np.ix_(keep_rows, keep_cols)
idx_bool
```

```python
A[idx_bool]
```

```python
idx_mixed = np.ix_([1, 3], keep_cols)
idx_mixed
```

```python
A[idx_mixed]
```

#### More on selecting rows and columns

```python
idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80, ['weight', 'origin']].head()
```

```python
indexer = lambda df: df['year'] > 80
Auto_re.loc[indexer, ['weight', 'origin']].head()
```

```python
indexer = lambda df: (df['year'] > 80) & (df['mpg'] > 30)
Auto_re.loc[indexer, ['weight', 'origin']].head()
```

```python
indexer = lambda df: (df['displacement'] < 300) & \
    (df.index.str.contains('ford')) | \
    (df.index.str.contains('datsun'))
Auto_re.loc[indexer, ['weight', 'origin', 'year']].head()
```
