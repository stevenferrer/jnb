```python
import pandas as pd
from matplotlib.pyplot import subplots
```

## College

```python
college = pd.read_csv('College.csv')
college.head()
```

```python
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis='columns')
college3 = college3.set_index('College')
college3.head()
```

```python
college = college3
college.head()
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college.loc[:, ['Top10perc', 'Apps', 'Enroll']]);
```

```python
fig, ax = subplots(figsize=(6, 6))
college.boxplot('Outstate', by='Private', ax=ax);
```

```python
college['Top10perc']
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
fig, ax = subplots(figsize=(6, 6))
college.boxplot('Outstate', by='Elite', ax=ax);
```

```python
college.describe()
```

```python
cols = ['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad', 'P.Undergrad', 'Grad.Rate']

fig, axes = subplots(nrows=4, ncols=2, figsize=(12, 12))

col_idx = 0
for i in range(0, 4):
    for j in range (0, 2):
        college.hist(cols[col_idx], bins=24, ax=axes[i, j])
        col_idx += 1
```
