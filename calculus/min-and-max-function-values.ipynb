```python
x, y, a = sp.symbols('x y a')

a_val = 1
# Define the equation
y_pos = sp.sqrt(x**3 / (2*a_val - x))
y_neg = -sp.sqrt(x**3 / (2*a_val - x))
y_pos, y_neg
```

```python
sp.plot(y_pos, y_neg, (x, 0.1, 1.9))
```
