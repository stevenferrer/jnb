```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np

import ipywidgets as widgets
from mpl_toolkits.mplot3d import axes3d

import sympy as sp
from sympy.plotting import plot_parametric

sp.init_printing()
```

```python
x = sp.symbols('x')
x
```

```python
f = 2*x**3 + 3*x**2 - 12*x + 1
f
```

```python
fp = sp.diff(f)
fp
```

```python
fpp = sp.diff(fp)
fpp
```

```python
fpp.subs(x, sp.Rational(-1,2))
```

```python
# find critical number
sp.solve(sp.Eq(fpp, 0))
```

```python
f.subs(x, sp.Rational(-1, 2))
```

```python
fp.subs(x, sp.Rational(-1, 2))
```

```python
fpp.subs(x, sp.Rational(-1, 2))
```

```python
sp.plot(f,fp,fpp, (x, -2, 2))
```

```python
for v in np.arange(-1.5, 1.5, .5):
    vals = [f.subs(x,v), fp.subs(x, v), fpp.subs(x, v)]
    print(vals)
```
